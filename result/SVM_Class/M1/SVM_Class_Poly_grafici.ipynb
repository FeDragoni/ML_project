```python
import pandas as pd
import numpy as np
from matplotlib import gridspec
import math
import seaborn as sns
# This file jupyter uses the results obtained with the grid search in order to plot the 
# results we have had with svm
```

```python
##RUN 1
```

```python
df = pd.read_csv("./M1_GS_poly_run_1.csv")
```

```python
df_new= df.drop(['Index'], axis=1)
df_new.sort_values("Score")
```

```python
df_new_1 = df_new.loc[df_new['degree'] == 2]
df_new_2 = df_new_1.loc[df_new['coef0'] == 0]
```

```python
df_new_2.sort_values("Score")
df_new_2 = df_new_2.sort_values(by = ["C", "gamma"])
df_new_2 = df_new_2.round(3)  ## approssima i dati a tot cifre decimali
# reset_index
df_new_2 = df_new_2.drop_duplicates(['C','gamma'])

dfp = df_new_2.pivot(index = "C", columns= "gamma", values = "Score")
ax = sns.heatmap(dfp, cmap="binary" )
```
