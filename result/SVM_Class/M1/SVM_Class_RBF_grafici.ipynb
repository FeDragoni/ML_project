```python
import pandas as pd
import numpy as np
from matplotlib import gridspec
import math
import seaborn as sns
# This file jupyter uses the results obtained with the grid search in order to plot the 
# results we have had with svm
```

```python
#####RUN 1 MONK 1
```

```python
df = pd.read_csv("./M1_GS_rbf_run1.csv")
```

```python
df_new= df.drop(['Index'], axis=1)
```

```python
df_new
```

```python
## per fare le TABELLINE!!
###per sortarlo rispetto ad un valore
df_new.sort_values("Score")
```

```python
###HEAT MAP !!!!!!
# df = pd.read_csv("/home/federico/Scrivania/ML_project-master/ML_project/result/SVM_CLASS_GS.csv")

df_new = df_new.sort_values(by = ["C", "gamma"])
df_new = df_new.round(3)  ## approssima i dati a tot cifre decimali
# reset_index
df_new = df_new.drop_duplicates(['C','gamma'])

dfp = df_new.pivot(index = "C", columns= "gamma", values = "Score")  #Return reshaped DataFrame 
                                                                       #organized by given index / column values.
ax = sns.heatmap(dfp, cmap="binary" )
```
