```python
import pandas as pd
import numpy as np
from matplotlib import gridspec
import math
import seaborn as sns
# This file jupyter uses the results obtained with the grid search in order to plot the 
# results we have had with svm
```

```python
#TRY FOR DEFINING C AND GAMMA LIMITATIONS
```

```python
df = pd.read_csv("./GS_poly_TRY_1.csv")
```

```python
df_new= df.drop(['Index'], axis=1)
```

```python
df.sort_values("Score")
```

```python
df_new = df_new.sort_values(by = ["estimator__C", "estimator__gamma"])
df_new = df_new.round(3)  ## approssima i dati a tot cifre decimali
# reset_index
df_new = df_new.drop_duplicates(['estimator__C','estimator__gamma'])

dfp = df_new.pivot(index = "estimator__C", columns= "estimator__gamma", values = "Score")  #Return reshaped DataFrame 
                                                                       #organized by given index / column values.
ax = sns.heatmap(dfp, cmap="binary" )
```
